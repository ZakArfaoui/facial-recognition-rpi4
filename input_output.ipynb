```python
import tensorflow.lite as tflite

MODEL_PATH = "./models/facenet.tflite"

itp = tflite.Interpreter(model_path=MODEL_PATH)
itp.allocate_tensors()

inp = itp.get_input_details()[0]
out = itp.get_output_details()[0]

print("INPUT :", inp["shape"], inp["dtype"])
print("OUTPUT:", out["shape"], out["dtype"])
print("IN quant :", inp.get("quantization"))
print("OUT quant:", out.get("quantization"))
```
